```python
import pandas as pd
import numpy as np
```

```python
Team1 = ["Joep", "Eef", "Kas", "Teun", "Tom", "Kenneth", "Tim P", "Tim H", "Lars", "Ken", "Mareno", 
        "Robin", "Marijn", "Stan", "Dylan", "Damjan"]

Team2 = ["Joep", "Eef", "Kas", "Teun", "Tom", "Kenneth", "Tim P", "Tim H", "Lars", "Ken", "Mareno", 
        "Robin", "Marijn", "Stan", "Dylan", "Damjan"]

Team3 = ["Joep", "Eef", "Kas", "Teun", "Tom", "Kenneth", "Tim P", "Tim H", "Lars", "Ken", "Mareno", 
        "Robin", "Marijn", "Stan", "Dylan", "Damjan"]

Data = ["ZONDAG 11 OKTOBER 2020", "ZONDAG 18 OKTOBER 2020", "ZONDAG 1 NOVEMBER 2020", "ZONDAG 22 NOVEMBER 2020", 
        "ZONDAG 13 DECEMBER 2020", "ZONDAG 24 JANUARI 2021", "ZONDAG 7 FEBRUARI 2021", "ZONDAG 7 MAART 2021",
       "ZONDAG 21 MAART 2021", "ZONDAG 11 APRIL 2021", "ZONDAG 25 APRIL 2021", "ZONDAG 9 MEI 2021"]

Tegenstander = ["Silvode", "VIOD", "Union", "Leones", "SC Bemmel",
                "Orion", "OBW", "VDZ 2", "Spero", "SML", "De Bataven", "Trekvogels"]
```

```python
# Make first dataframe with all players in three columns
df = pd.DataFrame(zip(Tegenstander, Data), 
                  columns=["Tegenstander", "Datum"])
df['Datum'] = df['Datum'].map(lambda x: x.lstrip('ZONDAG').rstrip('aAbBcC'))
df["Datum"] = df["Datum"].str.capitalize()
df2 = pd.DataFrame(zip(Team1, Team2, Team3), columns =["Speler 1", "Speler 2", "Speler 3"])
df2
```

```python
# Merge dataframe with dates and oponents
df3 = pd.concat([df,df2], ignore_index=True, axis=1)
df3.columns= ["Tegenstander", "Datum","Speler 1", "Speler 2", "Speler 3" ]
df3
```

```python
# Shuffle player columns
df4 = df3.copy() 
def shuffle(df4):     
    """Shuffles player columns and  makes sure that each player has to drive at least 2 times"""
    df4['True or False'] = np.where((df4["Speler 1"] == df4["Speler 2"]) | (df4["Speler 1"] == df4["Speler 3"]) | (df4["Speler 2"] == df4["Speler 3"]), 'True', 'False')
    while True:
        df4["Speler 1"] = df4["Speler 1"].sample(frac=1).reset_index(drop=True)
        df4["Speler 2"] = df4["Speler 2"].sample(frac=1).reset_index(drop=True)
        df4["Speler 3"] = df4["Speler 3"].sample(frac=1).reset_index(drop=True)
        df4['True or False'] = np.where((df4["Speler 1"] == df4["Speler 2"]) | (df4["Speler 1"] == df4["Speler 3"]) | (df4["Speler 2"] == df4["Speler 3"]), 'True', 'False')
        if 'True' not in df4['True or False'].values:  
            dff = df4[["Speler 1", "Speler 2", "Speler 3"]].copy()
            a = dff.apply(pd.Series.value_counts, axis=1).fillna(0)
            Total = []
            Total = a.iloc[[12,13,14,15],:].sum().tolist()
            Total = [int(i) for i in Total]
            while max(Total) < 1.5: # players cannot be seen twice after index 11 (when they do not have to drive)
                return df4
shuffle(df4)
```
